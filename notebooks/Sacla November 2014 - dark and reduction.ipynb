And then, let's retrieve corrections (it will take around 240s). To do this, we will use the `AnalysisProcessor` class. This class has been designed specifically to efficiently loop over images contained in SACLA data files. You can:

* preprocess each image (e.g. apply a ROI, apply a threshold)
* load and apply different analysis functions on each image (some basic functions are already available)
* provide user-defined functions

Given the fact that looping over the data file is expensive, each image is loaded in memory *once* and then all functions applied.

So, for example, let's see how to construct an `AnalysisProcessor` object, and what analysis are readily available:

```python
# name of the file
fname = DIR + "/full_hdf5/256635-257499/256635.h5"

# datasets to be analyzed. If the path is /run_XXXXX/dataset, then use just dataset (the run information is retrieved automatically)
dataset_names = ["detector_2d_1", "detector_2d_2"]

# create an AnalysisProcessor object
ap = ut.analysis.AnalysisProcessor()
# by default, the results will be returned in a dict, one entry per analysis
# if you do not want that, put this flag to True
ap.flatten_results = True

# list available analysis
print "Available preloaded analysis:", ap.available_analyses.keys()
# list available preprocess functions
print "Available preloaded preprocess functions:", ap.available_preprocess.keys()
# print help for a function
ap.print_help("image_get_spectra")
```

## Basic image analysis
What if we want to remove (subtract) the dark correction to all the images? We can then create a **preprocess function**, which must take an image as input, and return the processed image. 

```python
# That's a simple function
def subtract_dark(image, dark):
    return image - dark

fname = DIR + "/full_hdf5/256635-257499/256635.h5"

hf_dark = h5py.File(DIR + "/dark_corrections.h5", "r")
dark_1 = hf_dark["/run_256635/detector_2d_1/images_mean"][:]
dark_2 = hf_dark["/run_256635/detector_2d_2/images_mean"][:]

ap = ut.analysis.AnalysisProcessor()
ap.flatten_results = True
ap.add_analysis("image_get_mean_std")
ap.add_analysis("image_get_spectra", args={'axis': 1})
ap.add_analysis("image_get_histo_adu", args={"bins": np.arange(-100, 300, 5)})

# set the correct dataset (dark corrections are detector-dependent!). 
# Be careful that, as a security measure, this resets preprocess functions! If you do not want to, use remove_preprocess=False
ap.set_sacla_dataset(dataset_names[0])
# add the preprocess function
ap.add_preprocess(subtract_dark, args={'dark':dark_1})
# you can also use a lambda function, e.g.
#ap.add_preprocess(lambda image: image - results[0]["mean"], label="lambda_dark")

# to be safe, print the loaded preprocess and analysis functions:
print "Loaded analysis:", ap.list_analysis() 
print "Loaded preprocess functions:", ap.list_preprocess()

# run the new analysis!
results_dc_1 = ap.analyze_images(fname, n=-1)

# again, set up detector 2, and load the correct dark
ap.set_sacla_dataset(dataset_names[1])
ap.add_preprocess(subtract_dark, args={'dark':dark_2})

# and off we run!
results_dc_2 = ap.analyze_images(fname, n=-1)
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(131)
plt.plot(np.nansum(results_dc_1["spectra"], axis=0), label="detector 1")
plt.plot(np.nansum(results_dc_2["spectra"], axis=0), label="detector 2")
plt.legend(loc="best")
plt.subplot(132)
plt.bar(results_dc_1["histo_adu_bins"][:-1], results_dc_1["histo_adu"], width=5, log=True)
plt.subplot(133)
plt.bar(results_dc_2["histo_adu_bins"][:-1], results_dc_2["histo_adu"], width=5, log=True, color='b')
plt.tight_layout()
plt.show()
```

There are some striking features:

Let's focus on detector 2: what happens if we keep only the second peak ($ADU > 151$)?

```python
from scipy.signal import savgol_filter

ap.set_sacla_dataset(dataset_names[1])
ap.add_analysis("image_get_spectra", args={"axis": 1, }) 
ap.add_preprocess(subtract_dark, args={'dark':dark_2})
ap.add_preprocess("image_set_thr", args={'thr_low': 151})
results_data_2_156 = ap.analyze_images(fname, n=n_events)

spectra_1 = np.nansum(results_data_2["spectra"], axis=0)
spectra_2 = np.nansum(results_data_2_156["spectra"], axis=0)

plt.figure(figsize=(10,7))
plt.subplot(121)
plt.plot(spectra_1, label="detector 2 > 65")
plt.plot(spectra_2, label="detector 2 > 151")
plt.legend(loc="best", title=label)
plt.subplot(122)
plt.plot(savgol_filter(spectra_2 / spectra_1, 31, 3), label="smoothed efficiency")
plt.legend(loc="best")
plt.tight_layout()
plt.show()
```

The $ADU > 151$ cut on detector_2 seems to remove some diffuse scattering, leaving mostly inalterated the $K_\beta$ peak. 

Note that for smoothing the efficiency curve we used a filter from the `scipy.signal` package, which uses the [Savitzky-Golay](http://en.wikipedia.org/wiki/Savitzky%E2%80%93Golay_filter) filter

Excellent! We do see both $K_\alpha$ lines, and both $K_\beta$ and the elastic scattering peak, with a reduced diffuse background. 

So now, what if we want to:

* apply filters on data conditions (e.g. $I_0$), or
* do a laser on / laser off plot?

First of all, we need to get the quantities from the DAQ. As these files were produced with `DataConvert4`, these are already contained in the data file:

```python
# Define SACLA quantities - they can change from beamtime to beamtime
I0low = "event_info/bl_3/eh_4/photodiode/photodiode_I0_lower_user_7_in_volt"
I0up = "event_info/bl_3/eh_4/photodiode/photodiode_I0_upper_user_8_in_volt"
PDSample = "event_info/bl_3/eh_4/photodiode/photodiode_sample_PD_user_9_in_volt"
Mono = "event_info/bl_3/tc/mono_1_position_theta"
Delay = "event_info/bl_3/eh_4/laser/delay_line_motor_29"
ND = "event_info/bl_3/eh_4/laser/nd_filter_motor_26"
PhotonEnergy = "event_info/bl_3/oh_2/photon_energy_in_eV"
IsXray = "event_info/bl_3/eh_1/xfel_pulse_selector_status"
IsLaser = "event_info/bl_3/lh_1/laser_pulse_selector_status"
Tags = "event_info/tag_number_list"

run = "257681"
fh = h5py.File(DIR + "/hdf5/" + run + "_roi.h5", "r")
dset = fh["run_" + run]

# plotting
plt.figure(figsize=(10, 5))
plt.subplot(131)
plt.scatter(dset[I0low], dset[I0up])
plt.xlabel("$I_0^{low}$")
plt.ylabel("$I_0^{up}$")
plt.subplot(132)
plt.hist(dset[I0low], bins=np.arange(0, 1, 0.02), log=True)
plt.xlabel("$I_0^{low}$")
plt.subplot(133)
plt.hist(dset[I0up], bins=np.arange(0, 1, 0.02), log=True)
plt.xlabel("$I_0^{up}$")

plt.figure(figsize=(10, 5))
plt.subplot(131)
plt.hist(dset[PDSample], bins=np.arange(0, 0.6, 0.01), log=True)
plt.xlabel("PD signal")
plt.subplot(132)
plt.hist(dset[ND], log=True)
plt.xlabel("ND")
plt.subplot(133)
plt.hist(dset[PhotonEnergy], log=True)
plt.xlabel("Photon energy")

plt.tight_layout()
plt.show()
```

Let's define some cuts, to select only *good data*:

```python
is_data_mask = (dset[I0low][:] > 0.02) & (dset[I0up][:] > 0.02) & (dset[IsXray][:] == 1)
is_data_mask_2 = is_data_mask & (dset[I0low][:] < 0.5) & (dset[I0up][:] < 0.5)

spectra_1 = np.nansum(results_data_1["spectra"], axis=0)
spectra_1_isdata = np.nansum(results_data_1["spectra"][is_data_mask], axis=0)
spectra_1_isdata_2 = np.nansum(results_data_1["spectra"][is_data_mask_2], axis=0)

plt.figure()
plt.subplot(211)
plt.plot(spectra_1, label="Detector 1")
plt.plot(spectra_1_isdata, label="Detector 1, is_data")
plt.plot(spectra_1_isdata_2, label="Detector 1, is_data_2")
plt.legend(loc="best")
plt.subplot(212)
plt.plot(spectra_1_isdata / spectra_1, "o")
plt.plot(spectra_1_isdata_2 / spectra_1_isdata, "<")
plt.ylabel("Selection efficiency")

spectra_2 = np.nansum(results_data_2["spectra"], axis=0)
spectra_2_isdata = np.nansum(results_data_2["spectra"][is_data_mask], axis=0)
spectra_2_isdata_2 = np.nansum(results_data_2["spectra"][is_data_mask_2], axis=0)

plt.figure()
plt.subplot(211)
plt.plot(spectra_2, label="Detector 2")
plt.plot(spectra_2_isdata, label="Detector 2, is_data")
plt.plot(spectra_2_isdata_2, label="Detector 2, is_data_2")
plt.legend(loc="best")
plt.subplot(212)
plt.plot(spectra_2_isdata / spectra_2, "o")
plt.plot(spectra_2_isdata_2 / spectra_2_isdata, "<")
plt.ylabel("Selection efficiency")

plt.show()
```

Adding an higher threshold on the $I_0$ monitors seems only to reduce statistics, so it can be avoided at the moment. Selecting `laser_on` or `laser_off` is still very simple:

```python
laseron_mask = dset[IsLaser][:]

spectra_1_on = np.nansum(results_data_1["spectra"][is_data_mask & (laseron_mask == 1) ], axis=0)
spectra_1_off = np.nansum(results_data_1["spectra"][is_data_mask & (laseron_mask == 0 )], axis=0)

f, ax = plt.subplots(2, 1, sharex=True, )
ax[0].plot(spectra_1_on, label="On")
ax[0].plot(spectra_1_off, label="Off")
ax[0].legend(loc="best")
ax[1].plot(spectra_1_on - spectra_1_off, "b", label="on - off")
ax[1].plot(savgol_filter(spectra_1_on - spectra_1_off, 11, 3), "k", label="on - off smoothed", linewidth=2)
ax[1].legend(loc="best")
plt.tight_layout()
plt.show()
```

Starting from this point, it should be easy to extend the analysis in order to explore different cuts, or apply different analysis to the images.
