Now we can define some quantities, and apply some simple selections using just numpy:

````python
# Define SACLA quantities - they can change from beamtime to beamtime
daq_labels = {}
daq_labels["I0_down"] = "event_info/bl_3/eh_4/photodiode/photodiode_I0_lower_user_7_in_volt"
daq_labels["I0_up"] = "event_info/bl_3/eh_4/photodiode/photodiode_I0_upper_user_8_in_volt"
daq_labels["TFY"] = "event_info/bl_3/eh_4/photodiode/photodiode_sample_PD_user_9_in_volt"
daq_labels["photon_mono_energy"] = "event_info/bl_3/tc/mono_1_position_theta"
daq_labels["delay"] = "event_info/bl_3/eh_4/laser/delay_line_motor_29"
daq_labels["ND"] = "event_info/bl_3/eh_4/laser/nd_filter_motor_26"
daq_labels["photon_sase_energy"] = "event_info/bl_3/oh_2/photon_energy_in_eV"
daq_labels["x_status"] = "event_info/bl_3/eh_1/xfel_pulse_selector_status"
daq_labels["x_shut"] = "event_info/bl_3/shutter_1_open_valid_status"
daq_labels["laser_status"] = "event_info/bl_3/lh_1/laser_pulse_selector_status"
daq_labels["tags"] = "event_info/tag_number_list"

# loading some datasets, too
daq_datasets = {}
for k, v in daq_labels.iteritems():
    daq_datasets[k] = main_dset[v][:]

# one way of defining masks and selections:
x_status_mask = daq_datasets["x_status"]==1
x_shut_mask = daq_datasets["x_shut"]==1

# global mask. Here, & indicates the logical AND
mask = (x_status_mask) & (x_shut_mask)

# plotting
plt.figure()
plt.subplot(121)
plt.scatter(daq_datasets["tags"][mask], daq_datasets["I0_up"][mask])
plt.xlabel("tags")
plt.ylabel("$I_0^{up}$")
plt.subplot(122)
plt.hist(daq_datasets["I0_up"][mask], bins=np.arange(0, 1, 0.01), log=True)
plt.show()
````

## Using Pandas

As seen above, doing simple selections is pretty easy with Numpy. Nevertheless, if you need to do some complex selections or grouping, you can end up in writing long and complex code. In these cases, a tool as **Pandas** can become pretty useful. 

[Pandas](http://pandas.pydata.org/) is "an open source, BSD-licensed library providing high-performance, easy-to-use data structures and data analysis tools", and can provide a pretty simple and coherent view e.g. of experimental data and conditions. It is pretty useful also because it can load data in a large variety of formats (CSV, text, SQL, web pages...). It basically can sort data in a big, optimized table, and a lot of data mangling and statistical tools are already available for it.

The main Pandas entity is a `DataFrame`, which is a dataset composed of different `Series`: you can think of it as a sophisticated Excel table. Pandas DataFrame columns can be accessed both as a dict (`df["column_name"]`) and as an object (`df.column_name`)

Let's load our dataset in a `DataFrame`, and explore it a bit:

````python
# a temporary dictionary, where to store datasets
mydict = {}

t0 = 220.86

# Loading datasets
for k, v in daq_labels.iteritems():
    # assign to each dict entry an hdf5 dataset
    mydict[k] = main_dset[v][:].flatten()
    
    # operating some conversions on the fly
    if k == "delay":
        mydict[k] = sacla_converter.convert("delay", mydict[k], t0=t0)
    elif k == "photon_mono_energy":
        mydict[k] = sacla_converter.convert("energy", main_dset[v][:])
    elif k == "photon_sase_energy":
        mydict[k + "_mean"] = main_dset[v][:].mean()
    
# Create the dataframe
df = pd.DataFrame(mydict)

# print the first 10 entries
df.head(10)
````

The automatic index is a incrementing integer, but we can change that pretty easily:

````python
df = df.set_index("tags")
df.head()
````

We can apply selections on a DataFrame in a pretty straightforward way. For example, `df.x_status == 1` creates a boolean mask which can then be applied to `df` in order to select only the events where the *x ray status* was pinned as good. Multiple conditions can be concatenated with the `&` (AND) and `|` (OR) operators, e.g.:

```
is_data = (df.x_shut == 1) & (df.x_status == 1)
```

which gives now a mask where both the shutter and the status are ok. More complex selections can be performed... for example, if we want to select a particular delay, we can do:

```
new_df = df[df.delay == 2.0]
```
where we create a boolean mask (`df.delay == 2.0`) and immidiately apply it to the DataFrame.

Furthermore, we can easily extend the DataFrame with new colums, with a syntax similar to the one used for a `dict`:

````python
# adding two useful columns. & indicates the logical AND
df["is_data"] = (df['x_status'] == 1) & (df['x_shut'] == 1)
df["is_laser"] = (df['laser_status'] == 1)

# creating a simple mask
mask = df.is_data & df.is_laser

# plotting
plt.figure(figsize=(7, 5))
plt.subplot(121)
df[mask].I0_down.hist(bins=100, log=True)
plt.xlabel("I_0 down")

plt.subplot(122)
df[mask].I0_up.hist(bins=100, log=True)
plt.xlabel("I_0 up")

plt.figure(figsize=(7, 5))
(df.I0_up / df.I0_down).plot()
plt.xlabel("tags")
plt.ylabel("up / down")

plt.show()
````

As you can see, we used some special Pandas for plotting, like:

```
df[mask].I0_down.hist(bins=100, log=True)
```

This is just a wrapper of Matplotlib functions, and is equivalent to:
```
plt.hist(df[mask].I0_down, bins=100, log=True)
```
Both styles will be used, depending on the opportunity.

The advantage of using Pandas is that now we can load data from many files, and create coherent views from it, with indexes, grouping etc... Let's load data from various runs:

````python
runs = [ x for x in np.arange(259001, 259075)]
df_energy = pd.DataFrame(columns=["tags", "run", "photon_mono_energy", "I0_up"], )

tmp_data = {}
for run in runs:
    fname = DIR + str(run) +"_roi.h5" # the file name
    f = h5py.File(fname, "r")
    main_dset = f["run_" + str(run)]
    tmp_data["tags"] = main_dset[daq_labels["tags"]][:]
    tmp_data["run"] = run * np.ones(tmp_data["tags"].shape, dtype=int)
    tmp_data["I0_up"] = main_dset[daq_labels["I0_up"][:]]
    tmp_data["photon_mono_energy"] = sacla_converter.convert("energy", main_dset[daq_labels["photon_mono_energy"]][:])
    tmp_df = pd.DataFrame(data=tmp_data)
    df_energy = df_energy.append(tmp_df)
    
df_energy = df_energy.set_index('tags')

# getting the number of entries
df_energy.count()
````

Now we can do some interesting stuff... e.g., let's see the number of events per run. The `groupby` method returns a DataFrame grouped by a specific key:

````python
# counting events per run:
df_energy.groupby("run").count().I0_up.plot(kind="bar", legend=False, title="events per run", figsize=(7, 5))
plt.show()
# also as a pie... why not?
df_energy.groupby("run").count().I0_up.plot(kind="pie", legend=False, title="events per run", subplots=True, figsize=(5, 5))
plt.show()
````

Or, also, the Monochromator settings for a set of tags, or for runs. For example, `.loc` gives access to the possibility of slicing directly using the index values (the tag number in this case)

````python
# selecting a set of tags (which are the dataframe index now)
plt.subplot(221)
df_energy.loc[414139800:414139900].photon_mono_energy.plot(figsize=(7, 5))

# for a selected energy
plt.subplot(222)
df_energy[np.isclose(df_energy.photon_mono_energy, 9.67, rtol=0.001)].photon_mono_energy.plot(figsize=(7, 5))

# for a list of runs:
plt.subplot(223)
df_energy[df_energy.run.isin([259001, 259037])].photon_mono_energy.plot(figsize=(7, 5)) #logy=True)
plt.show()
````

## Computing XAS spectra

Let's have fun with a XAS plot... in principle, we do have all the information in the `event_info` dataset (see list above). it is enough then to sum up all the information from all the files. The needed steps are:

* Create a DataFrame
* Load the data, eventually performing some monipulation (conversions, etc)
* Add derived data, as the absorption coefficient

````python
# create a DataFrame
big_df = pd.DataFrame(columns=daq_datasets.keys(), )

for run in runs:
    mydict = {}  # temporary dict, where to store data
    fname = DIR + str(run) +"_roi.h5"  # the file name
    f = h5py.File(fname, "r")
    main_dset = f["run_" + str(run)]
    
    # Loading data from the specified datasets
    for k, v in daq_labels.iteritems():
        if k == "delay":
            # delays are in motor steps
            mydict[k] = sacla_converter.convert("delay", main_dset[v][:], t0=221)
        elif k == "photon_mono_energy":
            # mono energy settings are in motor steps
            mydict[k] = sacla_converter.convert("energy", main_dset[v][:])
        elif k == "photon_sase_energy":
            mydict[k + "_mean"] = main_dset[v][:].mean()
        else:
            mydict[k] = main_dset[v][:]
            
    # Append the data to the dataframe
    big_df = big_df.append(pd.DataFrame(data=mydict))

# round mono energy and delay
big_df.photon_mono_energy = np.round(big_df.photon_mono_energy.values, decimals=3)
big_df.delay = np.round(big_df.delay.values, decimals=2)

# create total I0 and absorption coefficients
big_df["I0"] = big_df.I0_up + big_df.I0_down
big_df["absorp"] = big_df.TFY / big_df.I0

# set tag number as index
big_df = big_df.set_index("tags")

# how many entries we do have?
print big_df.count()

# let's print the first 10 values
print big_df.head()

# let's plot an histogram with the I0 values
plt.figure()
plt.hist(big_df["I0"].values, bins=100)
plt.show()
````

Almost there... what we need to do now is:

* filter data
* create two different datasets for laser on and off
* perform a mean

All these steps are now pretty straightforward, as what we do have now is basically a huge table, with all the data we need.

Furthermore, Pandas DataFrames have plenty of *statistical functions*... for example, to get the overall mean, standard deviation and error of $I_0$ you can do:

````python
big_df.I0.mean(), big_df.I0.std(), big_df.I0.sem()
````

If, for example, we want instead the mean of all quantities in function of the Energy, we can either

* use the `groupby()` method, or
* reindex the DataFrame, and specify the `level` where to apply the mean (the `level` is just the name of the index you want to use for the mean... this is particularly helpful when you do have multi-indexes)

````python
big_df.groupby("photon_mono_energy").mean().head()
````

We do have now all the tools we need to produce the XAS plot:

````python
delay = 1.86

# preparing the is_data mask
is_data = (big_df.x_shut == 1) & (big_df.x_status == 1) & (big_df.photon_mono_energy > 9.6)
is_data = is_data & (big_df.I0_up > 0.01) & (big_df.I0_down > 0.01) & (big_df.ND > -1)

# filtering out garbage
data_df = big_df[is_data]

# select a particular delay (should not matter in this case...)
data_df = data_df[data_df.delay == delay]

# setting mono energy as index, as this is an energy scan:
data_df = data_df.set_index('photon_mono_energy')

# creating a laser_on/off dataframes
data_df_on = data_df[data_df.laser_status == 1]
data_df_off = data_df[data_df.laser_status == 0]

# plotting!
plt.figure()
df_on = data_df_on.absorp.mean(level=0)  # level 0 means "use the first index". This helps writing more generic code
df_off = data_df_off.absorp.mean(level=0)
df_on.plot(label="on")
df_off.plot(label="off")
(10 * (df_on - df_off)).plot(label="on - off (x10)")
plt.legend(loc="best", title="Delay=" + str(delay) + " ps")
plt.show()
````

Next piece of information we need is the standard error on the difference... also this is pretty easy to compute:

* get the standard error usign the `sem()` method
* combine the errors
* plot

````python
df_diff = (df_on - df_off)
# get standard error, using .sem() method
df_on_std = data_df[data_df.laser_status == 1].sem(level=0).absorp
df_off_std = data_df[data_df.laser_status == 0].sem(level=0).absorp
df_diff_std = np.sqrt((df_on_std) * (df_on_std) + (df_off_std) * (df_off_std))

### plot
fig = plt.figure()
# a global title (works also with subplots)
fig.suptitle("delay = %s ps" % (str(delay)), fontsize=20)
a2 = plt.subplot(111)
# plot the difference
df_diff.plot(label="on - off (w/ standard error)", color="k", linestyle="-", marker="o")
# plot a band with standard error
a2.fill_between(df_diff.index.tolist(), (df_diff - df_diff_std).values.tolist(),
                (df_diff + df_diff_std).values.tolist(), 
                alpha=0.8, # transparency value
                edgecolor='#999999', facecolor='#999999', # a bit of color...
                )
# y-label and legend (x-lable is generated automatically by pandas)
plt.ylabel("on - off")
plt.legend(title="", loc="best")
plt.show()
````

# Spanning through different scans

As an example, let's load data from different scans, and see how we can analyze it with Pandas. First, the usual loading:

````python
# create a DataFrame
runs = [ x for x in np.arange(259001, 259075)]
runs += [ x for x in np.arange(258825, 258883)]

big_df = pd.DataFrame(columns=daq_datasets.keys(), )

for run in runs:
    mydict = {}  # temporary dict, where to store data
    try:
        fname = DIR + str(run) +"_roi.h5"  # the file name
        f = h5py.File(fname, "r")
        main_dset = f["run_" + str(run)]
    except:
        pass
    
    # Loading data from the specified datasets
    for k, v in daq_labels.iteritems():
        if k == "delay":
            # delays are in motor steps
            mydict[k] = sacla_converter.convert("delay", main_dset[v][:], t0=220.86)
        elif k == "photon_mono_energy":
            # mono energy settings are in motor steps
            try:
                mydict[k] = sacla_converter.convert("energy", main_dset[v][:])
            except:
                print run, "conversion failed"
        elif k == "photon_sase_energy":
            mydict[k + "_mean"] = main_dset[v][:].mean()
        else:
            mydict[k] = main_dset[v][:]
            
    # Append the data to the dataframe
    big_df = big_df.append(pd.DataFrame(data=mydict))

# round mono energy and delay
big_df.photon_mono_energy = np.round(big_df.photon_mono_energy.values, decimals=3)
big_df.delay = np.round(big_df.delay.values, decimals=2)

# create total I0 and absorption coefficients
big_df["I0"] = big_df.I0_up + big_df.I0_down
big_df["absorp"] = big_df.TFY / big_df.I0

# set tag number as index
big_df = big_df.set_index("tags")
````

````python
# how many entries we do have?
print big_df.count()

# let's plot an histogram with the I0 values
plt.figure(figsize=(7,5))
big_df["I0"].hist(bins=100)
plt.show()
````

We can see how many events per delay setting:

````python
# let's plot an histogram with the I0 values
plt.figure(figsize=(7,5))
big_df[big_df.delay == 2.0].I0.hist(bins=100, label="delay=2 ps")
big_df[big_df.delay == 5.0].I0.hist(bins=100, label="delay=5 ps")
plt.legend(loc="best")
plt.show()
````
